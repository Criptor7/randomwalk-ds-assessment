```python
import matplotlib.pyplot as plt
```

```python
rating_buckets = [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0]
df['rating_bucket'] = pd.cut(df['average_rating'], bins=rating_buckets, labels=[f'{x:.1f} to {x + 0.4:.1f}' for x in rating_buckets[:-1]])
books_in_each_bucket = df.groupby('rating_bucket').size()
plt.bar(books_in_each_bucket.index, books_in_each_bucket.values)
plt.xlabel('Rating Bucket')
plt.ylabel('Number of Books')
plt.title('Number of Books in Each Rating Bucket')
plt.show()
```
